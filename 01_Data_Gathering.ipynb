```python
import json
```

```python
import os
tweets = []
for i in os.listdir('tweet/'):
    filename = 'tweet/' + i
    if filename.isdigit():
        with open(filename) as tweetfile:
            pyresponse = json.loads(tweetfile.read())
            tweets.append(pyresponse)
```

```python
tweets
```

```python
tweets = []
for file in os.listdir('tweet/'):
    filename = 'tweet/' + str(file)
    if filename[7:14].isdigit():
        with open(filename) as tweetfile:
            pyresponse = json.loads(tweetfile.read())
            tweets.append(pyresponse)
```

```python
tweets
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(tweets)
df.head()
```

```python
df.to_csv('tweets_scrape_first_df.csv')
```
