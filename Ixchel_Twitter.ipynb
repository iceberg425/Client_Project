```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
from twitterscraper import query_tweets
import time
```

```python
data = []
for tweet in query_tweets('conedison%20"power%20outage"%20since%3A2017-01-01%20until%3A2017-07-28&src=typd', 10):
    print(tweet.text)
    tweets = {}
    tweets['userid'] = tweet.user.encode('utf-8')
    tweets['id'] = tweet.id
    tweets['date'] = tweet.timestamp
    tweets['tweet'] = tweet.text
    data.append(tweets)
    time.sleep(2)
    
df = pd.DataFrame(data)
```

```python
import pandas as pd
df = pd.DataFrame(data)
```

```python
df.to_csv('7.28.17_to_7_01_17.csv')
```
